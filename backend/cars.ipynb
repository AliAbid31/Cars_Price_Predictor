```python
urban_types = ['SUV', 'Compacte', 'Citadine', 'Berline', 'Monospace']
utility_types = ['Pickup', 'Utilitaire'] 
```

```python
df_urban = cars_cleaned[cars_cleaned['Car Type'].isin(urban_types)].copy()
df_utility = cars_cleaned[cars_cleaned['Car Type'].isin(utility_types)].copy()
```

```python
def train_specialist_model(df_segment):
    """
    Prend un segment du DataFrame, le prépare, et entraîne un modèle XGBoost.
    Retourne le modèle entraîné, le scaler, et la liste des colonnes utilisées.
    """
    # Préparation spécifique au segment
    segment_encoded = pd.get_dummies(df_segment, drop_first=True)
    
    X = segment_encoded.drop('Price', axis=1)
    y_log = np.log1p(segment_encoded['Price']) # On utilise la transformation log !
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Entraînement du modèle sur 100% des données du segment
    model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
    model.fit(X_scaled, y_log)
    
    print(f"Modèle entraîné pour un segment de {len(df_segment)} lignes.")
    
    return model, scaler, X.columns
```

```python
urban_model, urban_scaler, urban_columns = train_specialist_model(df_urban)
utility_model, utility_scaler, utility_columns = train_specialist_model(df_utility)
```

```python
models_dict = {
    'urban': {'model': urban_model, 'scaler': urban_scaler, 'columns': urban_columns},
    'utility': {'model': utility_model, 'scaler': utility_scaler, 'columns': utility_columns}
}
```

```python
def predict_price_master(new_car_data):
    """
    Prend les données d'une nouvelle voiture (sous forme de dictionnaire),
    choisit le bon modèle spécialiste, et retourne le prix prédit.
    """
    # On transforme les données en DataFrame d'une seule ligne
    new_car_df = pd.DataFrame([new_car_data])
    car_type = new_car_df.iloc[0]['Car Type']
    
    # --- ROUTAGE : Choisir le bon spécialiste ---
    segment = None
    if car_type in urban_types:
        segment = 'urban'
    elif car_type in utility_types:
        segment = 'utility'
    
    if segment is None:
        return f"Erreur: Le type de voiture '{car_type}' n'est pas géré."
        
    print(f"Routage vers le modèle spécialiste : '{segment}'")
    
    # --- Préparation des données pour le modèle choisi ---
    specialist = models_dict[segment]
    model = specialist['model']
    scaler = specialist['scaler']
    columns = specialist['columns']
    
    # Appliquer le même One-Hot Encoding
    new_car_encoded = pd.get_dummies(new_car_df)
    # Très important : s'assurer que le DataFrame a les mêmes colonnes que lors de l'entraînement
    new_car_reindexed = new_car_encoded.reindex(columns=columns, fill_value=0)
    
    # Appliquer le même Scaling
    new_car_scaled = scaler.transform(new_car_reindexed)
    
    # --- Prédiction ---
    prediction_log = model.predict(new_car_scaled)
    
    # Re-transformer la prédiction log en prix normal
    final_prediction = np.expm1(prediction_log)
    
    return final_prediction[0]
```

```python
models_zoo = {}
# Entraîner les 2 modèles par type de véhicule
models_zoo['urban'] = train_specialist_model(df_urban)
models_zoo['utility'] = train_specialist_model(df_utility)
```

```python
models_zoo['level1'] = train_specialist_model(df_level1)
models_zoo['level2'] = train_specialist_model(df_level2)
models_zoo['level3'] = train_specialist_model(df_level3)
models_zoo['level4'] = train_specialist_model(df_level4)
models_zoo['level5'] = train_specialist_model(df_level5)
```

```python
def predict_price_master(new_car_data, prediction_mode='level'):
    """
    Prend les données d'une nouvelle voiture et un mode de prédiction.
    Choisit le bon modèle spécialiste et retourne le prix prédit.
    prediction_mode peut être 'level' (par défaut) ou 'type'.
    """
    new_car_df = pd.DataFrame([new_car_data])
    
    # --- ROUTAGE : Choisir le bon spécialiste ---
    model_key = None
    if prediction_mode == 'type':
        car_type = new_car_df.iloc[0]['Car Type']
        if car_type in urban_types: model_key = 'urban'
        elif car_type in utility_types: model_key = 'utility'
    else: # Par défaut, on utilise les modèles par niveau
        brand = new_car_df.iloc[0]['Brand']
        level = assigner_level(brand)
        if level > 0: model_key = f'level{level}'

    if model_key is None or model_key not in models_zoo:
        return f"Erreur: Impossible de trouver un modèle pour cette voiture avec le mode '{prediction_mode}'."
        
    print(f"Routage vers le modèle spécialiste : '{model_key}'")
    
    # --- Préparation des données pour le modèle choisi ---
    model, scaler, columns = models_zoo[model_key]
    
    # On doit appliquer les mêmes transformations que pour l'entraînement
    new_car_df['Level'] = new_car_df['Brand'].apply(assigner_level) # Ajouter le level
    new_car_df = new_car_df.drop(columns=['Brand', 'Model', 'Ext Color', 'Car Type', 'Motorisation', 'Gearbox type'], errors='ignore')
    
    new_car_encoded = pd.get_dummies(new_car_df)
    new_car_reindexed = new_car_encoded.reindex(columns=columns, fill_value=0)
    new_car_scaled = scaler.transform(new_car_reindexed)
    
    # --- Prédiction ---
    prediction_log = model.predict(new_car_scaled)
    final_prediction = np.expm1(prediction_log)
    
    return final_prediction[0]
```

```python
def train_specialist_pipeline(df_segment):
    """
    Prend un segment de DataFrame, construit un pipeline complet
    (preprocessing + modèle), l'entraîne, et le retourne.
    """
    # Exclure les colonnes qui ne servent plus directement
    features_to_drop = ['Brand', 'Model', 'Car Type']
    df_segment = df_segment.drop(columns=features_to_drop, errors='ignore')

    X = df_segment.drop('Price', axis=1)
    y_log = np.log1p(df_segment['Price'])
    
    # Définir les colonnes à transformer
    numeric_features = X.select_dtypes(include=np.number).columns
    categorical_features = X.select_dtypes(include='object').columns
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numeric_features),
            ('cat', OneHotEncoder(handle_unknown='ignore', drop='first'), categorical_features)
        ], remainder='passthrough')

    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42))
    ])
    
    pipeline.fit(X, y_log)
    print(f"Pipeline entraîné pour un segment de {len(df_segment)} lignes.")
    return pipeline
```

```python
def train_and_evaluate_specialist(df_segment, segment_name):
    """
    Prend un segment de DataFrame et son nom :
    1. Le divise en train/test.
    2. Entraîne un pipeline sur le train set.
    3. Évalue le score R² sur le test set et l'affiche.
    4. Ré-entraîne le pipeline sur 100% des données du segment.
    5. Retourne le pipeline final prêt pour la production.
    """
    print(f"\n--- Traitement du segment : {segment_name} ({len(df_segment)} lignes) ---")
    
    # Sécurité : si le segment est trop petit, on ne peut pas l'entraîner
    if len(df_segment) < 10:
        print("Segment trop petit pour l'entraînement. Modèle non créé.")
        return None

    # 1. Séparation des données du segment
    features_to_drop = ['Brand', 'Model', 'Car Type']
    df_segment_processed = df_segment.drop(columns=features_to_drop, errors='ignore')
    
    X = df_segment_processed.drop('Price', axis=1)
    y_log = np.log1p(df_segment_processed['Price'])
    
    X_train, X_test, y_train_log, y_test_log = train_test_split(
        X, y_log, test_size=0.2, random_state=42
    )

    # 2. Création du pipeline (identique à avant)
    numeric_features = X.select_dtypes(include=np.number).columns
    categorical_features = X.select_dtypes(include='object').columns
    
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numeric_features),
            ('cat', OneHotEncoder(handle_unknown='ignore', drop='first'), categorical_features)
        ], remainder='passthrough')

    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42))
    ])
    
    # 3. Entraînement sur le jeu d'entraînement du segment
    pipeline.fit(X_train, y_train_log)
    
    # 4. ÉVALUATION sur le jeu de test du segment
    score = pipeline.score(X_test, y_test_log)
    print(f"✅ Score R² pour le segment '{segment_name}' : {score:.4f}")

    # 5. Ré-entraînement sur 100% des données pour le modèle final
    print("Ré-entraînement du modèle sur 100% des données du segment...")
    pipeline.fit(X, y_log)
    
    return pipeline
```

```python
# On crée notre "zoo" de pipelines
pipelines_zoo = {}

# On crée un dictionnaire des segments pour boucler dessus proprement
segments = {
    'urban': df_urban,
    'utility': df_utility,
    'level1': df_level1,
    'level2': df_level2,
    'level3': df_level3,
    'level4': df_level4,
    'level5': df_level5
}
```
